# **JSON based Inventory Management System**

```python
import json
import os
import time
fd=open('Inventory.json','r')
r=fd.read()
fd.close()
record = json.loads(r)
def menu():
    print('*********GK_MOBILES**********')
    print('=============================')
    print('= Inventory Management Menu =')
    print('=============================')
    print('(1) Print Inventory Report')
    print('(2) Add New mobile to Inventory')
    print('(3) Purchase new mobile')
    print('(4) Remove a device from Inventory')
    print('(5) Search a device in Inventory')
    print('(6) Check for stock availability')
    print('(0) Quit')
    c = int(input("Enter choice: "))
    menuSelection(c)
def menuSelection(c):
    if c == 1:
        displayInventory()
    elif c == 2:
        addInventory()
    elif c==3:
        purchase()
    elif c==4: 
        removeInventory()
    elif c==5:
        searchInventory()
    elif c==6:
         stock()               
    elif c == 0:
        exit()
def stock():
    d_id=input("Enter the device id:")
    if (record[d_id]['quantity']==0):
      print("No stocks available for this device id")
    else:
      print("Stocks available")
    c = int(input('Enter 1 to continue or 0 to exit: '))
    if c == 1:
          menu()
    else:
          exit()
def purchase():
    print("==========Bill Receipt=========")
    print("==========GK_Mobiles===========")
    print("*******************************")
    u_d_id=input("Enter the device id:")
    u_quant=int(input("Enter the quantity:"))
    print("*******************************")
    print("Device id:",u_d_id)
    print("Brand    :",record[u_d_id]['brand'])
    print("Model     :",record[u_d_id]['model'])
    print("Quantity  :",u_quant)
    print("Price     :",record[u_d_id]['price'])
    print("*******************************")
    print("Total amount:",record[u_d_id]['price'] * u_quant)
    print("Purchased time:",time.ctime())
    print("*******************************")
    print("Thanks for Coming")
    print("Please come again")
    record[u_d_id]['quantity'] -= u_quant
    js = json.dumps(record)
    fd = open("Inventory.json",'w')
    fd.write(js)
    fd.close()
    sales={'Device_id':u_d_id,'Brand':record[u_d_id]['brand'],'Model':record[u_d_id]['model'],'Quantity':u_quant,'Price':record[u_d_id]['price'],'NetAmount':record[u_d_id]['price'] * u_quant,'Purchased time':time.ctime()}
    jd=json.dumps(sales)
    fs=open('Sales.json','a')
    fs.write(jd)
    fs.close()
    c = int(input('Enter 1 to continue or 0 to exit: '))
    if c == 1:
          menu()
    else:
          exit()
def removeInventory():
  fd = open('Inventory.json', 'w')
  print("Removing an existing device from Inventory")
  print("=========================================")
  d_id = input("Enter the device id to remove from inventory: ")
  del record[d_id]
  js = json.dumps(record)
  fd.write(js)
  fd.close()
  c = int(input('Enter 1 to continue or 0 to exit: '))
  if c == 1:
    menu()
  else:
    exit()
def searchInventory():
  print('Searching the device from the Inventory')
  print('===================')
  s_id = input("Enter the device id to search from inventory: ")
  print("The device found with the entered id :")
  print(record[s_id])
  CHOICE = int(input('Enter 1 to continue or 0 to exit: '))
  if CHOICE == 1:
    menu()
  else:
    exit()             
def displayInventory():
  print(record) 
  x = int(input('Enter 1 to continue or 0 to exit: '))
  if x == 1:
          menu()
  else:
          exit()        
def addInventory():
      print("Adding mobile to Inventory")
      print("==========================")
      device_id=input("Enter the id of the mobile:")
      brand = input("Enter the name of the mobile: ")
      model=input("Enter the device model:")
      quantity =int(input("Enter the quantity of the mobile: "))
      price=int(input("Enter the price of the mobile:"))
      record[device_id]={'brand':brand,'model':model,'quantity':quantity,'price':price}
      j=json.dumps(record)
      fd = open('Inventory.json', 'w')
      fd.write(j)
      fd.close()
      c = int(input('Enter 1 to continue or 0 to exit: '))
      if c == 1:
          menu()
      else:
          exit()   
menu()
```
